```python
import collections
import sys
import time
import heapq
class puzzle_solver:

    def __init__(self,initial_state,goal_state):
        self.initial_state=self.convert_to_list(initial_state)
        self.goal_state=self.convert_to_list(goal_state)
        self.number_of_visited_nodes=0
        self.path_cost=0
        self.memory_consumed=0
        self.time_taken=0.0
        self.path=[]
        self.visited=set()
        self.parent={}


    def bfs(self):
        print("bfs")
        start_time = time.time()
        queue = collections.deque()
        queue.append(self.initial_state)
        self.visited.add(tuple(self.initial_state))
        max_queue_size = 0
        while queue:
            max_queue_size = max(max_queue_size, len(queue))
            current_state = queue.popleft()
            self.number_of_visited_nodes += 1
            if current_state == self.goal_state:
                print("Goal reached!")
                self.print_results()
                self.print_matrix(current_state)
                break
            zero_index = current_state.index(0)
            row, col = zero_index // 3, zero_index % 3
            moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]
            for move in moves:
                new_row, new_col = row + move[0], col + move[1]
                if 0 <= new_row < 3 and 0 <= new_col < 3:
                    new_index = self.convert_index_to_position(new_row, new_col)
                    new_state = current_state.copy()
                    new_state[zero_index], new_state[new_index] = new_state[new_index], new_state[zero_index]
                    if tuple(new_state) not in self.visited:
                        self.visited.add(tuple(new_state))
                        self.parent[tuple(new_state)] = tuple(current_state)
                        queue.append(new_state)
        self.time_taken=time.time()-start_time
        self.memory_consumed = max_queue_size
        print("Memory Consumed: ",self.memory_consumed)
        print("Time  taken:",self.time_taken)
        print("____________________")
        pass

    def print_matrix(self,state):
        for i in range(3):
            for j in range(3):
                print(state[i*3+j],end=" ")
            print()


    def dfs(self):
        print("Dfs")
        start_time = time.time()
        stack = collections.deque()
        stack.append(self.initial_state)
        self.visited.add(tuple(self.initial_state))
        max_stack_size = 0
        while stack:
            max_stack_size = max(max_stack_size, len(stack))
            current_state = stack.pop()
            self.number_of_visited_nodes += 1
            if current_state == self.goal_state:
                print("Goal reached!")
                self.print_results()
                self.print_matrix(current_state)
                break
            zero_index = current_state.index(0)
            row, col = zero_index // 3, zero_index % 3
            moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]
            for move in moves:
                new_row, new_col = row + move[0], col + move[1]
                if 0 <= new_row < 3 and 0 <= new_col < 3:
                    new_index = self.convert_index_to_position(new_row, new_col)
                    new_state = current_state.copy()
                    new_state[zero_index], new_state[new_index] = new_state[new_index], new_state[zero_index]
                    if tuple(new_state) not in self.visited:
                        self.visited.add(tuple(new_state))
                        self.parent[tuple(new_state)] = tuple(current_state)
                        stack.append(new_state)
        self.time_taken=time.time()-start_time
        self.memory_consumed =max_stack_size
        print("Memory Consumed: ",self.memory_consumed)
        print("Time  taken:",self.time_taken)
        print("_____________________")
        pass

    def ucs(self):
        print("Ucs")
        start_time = time.time()
        priority_queue = [(0, self.initial_state)]
        heapq.heappush(priority_queue, (0, self.initial_state))
        self.visited.add(tuple(self.initial_state))
        max_queue_size = 0
        while priority_queue:
            max_queue_size = max(max_queue_size, len(priority_queue))
            cost,current_state = heapq.heappop(priority_queue)
            self.number_of_visited_nodes += 1
            if current_state == self.goal_state:
                print("Goal reached!")
                self.print_results()
                print("The Cost is :",cost)
                self.print_matrix(current_state)
                break
            zero_index = current_state.index(0)
            row, col = zero_index // 3, zero_index % 3
            moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]
            for move in moves:
                new_row, new_col = row + move[0], col + move[1]
                if 0 <= new_row < 3 and 0 <= new_col < 3:
                    new_index = self.convert_index_to_position(new_row, new_col)
                    new_state = current_state.copy()
                    if new_state==self.goal_state:
                      print("Goal reached!")

                    new_state[zero_index], new_state[new_index] = new_state[new_index], new_state[zero_index]
                    if tuple(new_state) not in self.visited:
                        self.visited.add(tuple(new_state))
                        self.parent[tuple(new_state)] = tuple(current_state)
                        heapq.heappush(priority_queue, (cost + 1, new_state))
        self.time_taken=time.time()-start_time
        self.memory_consumed =max_queue_size
        print("Memory Consumed: ",self.memory_consumed)
        print("Time  taken:",self.time_taken)
        print("_____________________")
        pass





    def convert_index_to_position(self,row,col):
        return row*3+col
    def a_star(self):
        pass
    def print_results(self):
        self.get_path()
        self.get_number_of_nodes_visited()
        self.get_path_cost()


        pass

    def get_number_of_nodes_visited(self):
        print("Number of nodes visited: ",self.number_of_visited_nodes)
        pass

    def get_path_cost(self):
        pass

    def get_memory_consumed(self):
        print("Memory consumed: ",self.memory_consumed)
        pass
    def get_time_taken(self):
      print("Time Taken : ",self.time_taken)
      return

    def get_path(self):

        path = []
        current_state = tuple(self.goal_state)
        while current_state != tuple(self.initial_state):
            path.append(current_state)
            current_state = self.parent.get(current_state)
            if current_state is None:
                print("No path found")
                return []
        path.append(tuple(self.initial_state))
        path.reverse()
        print(path)
        return path

    def convert_to_string(self):
        pass

    def convert_to_list(self,s):
        l=[]
        for each in s:
            l.append(int(each))
        return l


    def ids(self, max_depth):
        print("IDS Starting")
        start_time = time.time()
        max_memory_consumed=0
        for limit in range(1, max_depth + 1):
            self.visited.clear()
            self.parent.clear()
            self.visited.add(tuple(self.initial_state))
            stack = [(self.initial_state, limit)]
            while stack:
                max_memory_consumed = max(max_memory_consumed, len(stack))
                current_state, current_limit = stack.pop()
                if current_state == list(self.goal_state):
                    self.print_matrix(current_state)
                    print("Goal reached!")
                    print(f"Time taken: {time.time() - start_time} seconds")
                    print("Nodes Visited: ",self.number_of_visited_nodes)
                    print("Max Memory Consumed:",max_memory_consumed)
                    return
                if current_limit > 0:
                    zero_index = current_state.index(0)
                    row, col = zero_index // 3, zero_index % 3
                    moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]
                    for move in moves:
                        new_row, new_col = row + move[0], col + move[1]
                        if 0 <= new_row < 3 and 0 <= new_col < 3:
                            new_index = self.convert_index_to_position(new_row, new_col)
                            new_state = current_state.copy()
                            new_state[zero_index], new_state[new_index] = new_state[new_index], new_state[zero_index]
                            if tuple(new_state) not in self.visited:
                                self.number_of_visited_nodes += 1
                                self.visited.add(tuple(new_state))
                                self.parent[tuple(new_state)] = tuple(current_state)
                                stack.append((new_state, current_limit - 1))
        print("Goal not found within depth limit.")
        return False


bfs=puzzle_solver("120345678","012345678")
bfs.bfs()
dfs=puzzle_solver("120345678","012345678")
dfs.dfs()
ucs=puzzle_solver("120345678","012345678")
ucs.ucs()
ids=puzzle_solver("120345678","012345678")
ids.ids(50)
path=ids.get_path()
print("Path Cost: ",len(path))
```
